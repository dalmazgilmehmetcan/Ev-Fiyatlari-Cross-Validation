```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Veri setimizi yüklüyoruz.
veriler = pd.read_csv('ev-fiyat-verisi.csv')
veriler.head()
```

```python
# Verimiz içerisindeki numerik ve kategorik değerleri ayırıyoruz.
sehir = veriler.iloc[:,0:1].values # kategorik
numeric_veri = veriler.iloc[:,1:].values
```

```python
# Kategorik değer olan "şehir" değerine sırayla label ve one hot encoder uyguluyoruz.
le = LabelEncoder()
sehir[:,0] = le.fit_transform(sehir[:,0])

ohe = OneHotEncoder()
sehir=ohe.fit_transform(sehir).toarray()
```

```python
# Verimizi numpy dizisinden pandas dataframe'i haline getiriyoruz.
sehir_df = pd.DataFrame(data = sehir, index = range(22), columns=['ankara','bursa','istanbul'] )
numeric_veri_df =pd.DataFrame(data = numeric_veri, index = range(22), columns = ['oda_sayisi','salon_sayisi','ev_yasi','fiyat'])

#dataframe birlestirme islemi
veri_birlestirme=pd.concat([sehir_df,numeric_veri_df],axis=1)
veri_birlestirme
```

```python
# Verimizi bağımsız ve bağımlı değişken olarak ayırıyoruz.
x=veri_birlestirme.iloc[:,0:6]
y=veri_birlestirme.iloc[:,6:]
X=x.values
Y=y.values
```

```python
# Lineer Regresyon
lr=LinearRegression()
lr.fit(X,Y)
y_pred_lr = lr.predict(X)

print(f"Lineer Regresyon r^2 score = {r2_score(Y,y_pred_lr)}")
```

```python
# Polinom
poly_reg = preprocessing.PolynomialFeatures()
x_poly=poly_reg.fit_transform(X)

lr2=LinearRegression()
lr2.fit(x_poly,Y)
Y_pred_poly=lr2.predict(x_poly)

print(f"Polinom r^2 score = {r2_score(Y,Y_pred_poly)}")
```

```python
# StandardScale
sc = preprocessing.StandardScaler()
X_olc=sc.fit_transform(X)
y_train_olc=sc.fit_transform(Y)

# Support Vector Machine
svr_reg = SVR(kernel='rbf')
svr_reg.fit(X_olc,y_train_olc)
y_pred_svr = svr_reg.predict(X_olc)

print(f"SVR r^2 score = {r2_score(y_train_olc,y_pred_svr)}")
```

```python
# Karar Ağacı
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X,Y)
y_pred_dt = dtr.predict(X)

print(f"Karar Ağacı r^2 score = {r2_score(Y,y_pred_dt)}")
```

```python
# Random Forest
rf_reg = RandomForestRegressor(random_state=0,n_estimators=15)
rf_reg.fit(X,Y.ravel())
y_pred_rf=rf_reg.predict(X)

print(f"Random Forest r^2 score = {r2_score(Y, y_pred_rf)}")
```
